```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
```

```python
name_model = "mistralai/Mistral-7B-v0.1"
promt = ["the list colors: red, blue, green"]


quantization_config = BitsAndBytesConfig(load_in_4bit=True)
model = AutoModelForCausalLM.from_pretrained(name_model, device_map="auto", quantization_config=quantization_config)
```

```python
prompt = ["a list of animal: cat, dog, fish"]
```

```python
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-v0.1", padding_side="left")
model_inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
```

```python
generated_ids = model.generate(**model_inputs)
tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
```
