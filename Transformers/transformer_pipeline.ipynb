```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
# classifier("I've been waiting for a HuggingFace course my whole life.")
classifier(["I like dog.","My brother hate cat"])
```

# Zero short classification

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "I usually play chess in the morning.",
    candidate_labels=["sport", "routine", "business"],
)
```

# Test generation

```python
generator = pipeline("text-generation")
generator("I want pet ownership and cat and dog", max_length = 15)
```

# Using model in pipeline

```python
generator = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M")
generator(
    "At school, teacher will teach you",
    max_length=30,
    num_return_sequences=2,
)
```
